# 03 — Model Benchmark (Walk-Forward CLV)

Compare **17 models** on 5 temporal walk-forward windows using Normalized Gini, MAE, R², and revenue capture @ top 10%.

Primary table: `results/tables/main_model_comparison.csv`  
Full pipeline: `src/models_baselines.py` → … → `src/final_model_comparison.py`

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
TABLE = ROOT / "results" / "tables" / "main_model_comparison.csv"
FALLBACK = ROOT / "results" / "MASTER_TABLE.csv"

path = TABLE if TABLE.exists() else FALLBACK
df = pd.read_csv(path)
print(f"Loaded {len(df)} models from {path.name}")
assert len(df) >= 5, "ADY201m requires at least 5 models"
df
```

## Model categories

```python
if "Category" in df.columns:
    display(df.groupby("Category").size().rename("count").sort_values(ascending=False))
else:
    print("Categories not tagged — see Model column")
```

## Top models by Normalized Gini (mean ± std)

```python
rank = df.sort_values("Norm_Gini_mean", ascending=False)
cols = ["Model", "Norm_Gini_mean", "Norm_Gini_std", "MAE_mean", "Revenue_Capture_10_mean", "Category"]
display(rank[cols].head(10))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
top = df.nlargest(12, "Norm_Gini_mean").sort_values("Norm_Gini_mean")
ax.barh(top["Model"], top["Norm_Gini_mean"], xerr=top["Norm_Gini_std"], color="#2E86AB", capsize=3)
ax.set_xlabel("Normalized Gini (mean ± std)")
ax.set_title("Walk-forward model comparison — top 12 by Gini")
plt.tight_layout()
plt.show()
```
