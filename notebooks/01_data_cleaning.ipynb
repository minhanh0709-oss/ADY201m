# 01 — Data Cleaning & Preprocessing

**ADY201m pipeline step 1:** load Online Retail II, clean transactions, and export processed data for feature engineering.

| Step | Action |
|------|--------|
| Load | Merge Year 2009–2010 + 2010–2011 sheets |
| Clean | Drop missing CustomerID, negative qty/price, duplicates |
| Filter | Remove cancelled invoices (InvoiceNo starting with `C`) |
| Export | `data/processed/online_retail_cleaned.csv` |

Full implementation: `src/data_cleaning.py`

```python
import subprocess
import sys
from pathlib import Path

import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
RAW = ROOT / "data" / "raw" / "online_retail_II.xlsx"
CLEANED = ROOT / "data" / "processed" / "online_retail_cleaned.csv"
SAMPLE = ROOT / "data" / "processed" / "sample_window_3_features.csv"

print("Project root:", ROOT)
print("Raw file exists:", RAW.exists())
print("Cleaned file exists:", CLEANED.exists())
```

## Run cleaning script

Requires `data/raw/online_retail_II.xlsx` (see `data/raw/README.md`). Skip if already cleaned.

```python
if RAW.exists() and not CLEANED.exists():
    subprocess.run([sys.executable, str(ROOT / "src" / "02_data_cleaning.py")], check=True)
elif not RAW.exists():
    print("Download raw data first — see data/raw/README.md")
else:
    print("Using existing cleaned file:", CLEANED)
```

## Inspect cleaned data

```python
if CLEANED.exists():
    df = pd.read_csv(CLEANED, parse_dates=["InvoiceDate"], nrows=5000)
    print(f"Shape (sample): {df.shape}")
    print(f"Date range: {df['InvoiceDate'].min()} → {df['InvoiceDate'].max()}")
    print(f"Unique customers (sample): {df['CustomerID'].nunique():,}")
    display(df.head())
    display(df.describe(include="all").T.head(10))
elif SAMPLE.exists():
    print("Full cleaned CSV not local — showing committed sample features:")
    display(pd.read_csv(SAMPLE).head())
```
